```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def mostrar_sistema_matricial(a, b, c, d):
    A = np.array([[a, b], [c, d]])
    print("Sistema de ecuaciones en formato matricial:")
    print("d/dt [ x ] = [ a  b ] [ x ]")
    print("      [ y ]   [ c  d ] [ y ]")
    print("Matriz de coeficientes:")
    print(A)
    
    # Cálculo de los valores propios
    discriminante = (a + d)**2 - 4 * (a*d - b*c)
    if abs(discriminante) < 1e-20:
        discriminante = 0
    
    if discriminante >= 0:
        lambda1 = ((a + d) + np.sqrt(discriminante)) / 2
        lambda2 = ((a + d) - np.sqrt(discriminante)) / 2
    else:
        lambda1 = complex((a + d) / 2, np.sqrt(-discriminante) / 2)
        lambda2 = complex((a + d) / 2, -np.sqrt(-discriminante) / 2)
    
    # Graficar
    t = np.linspace(0, 10, 100)
    dx_dt = np.exp(lambda1.real * t)
    dy_dt = np.exp(lambda2.real * t)
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Diagrama de fases
    x_vals = np.linspace(-10, 10, 20)
    y_vals = np.linspace(-10, 10, 20)
    X, Y = np.meshgrid(x_vals, y_vals)
    U = a * X + b * Y
    V = c * X + d * Y
    
    axes[0].quiver(X, Y, U, V, color='blue')
    axes[0].axvline(x=lambda1.real, color='r', linestyle='--', label=f"λ₁ = {lambda1}")
    axes[0].axhline(y=lambda2.real, color='g', linestyle='--', label=f"λ₂ = {lambda2}")
    axes[0].set_xlabel('x')
    axes[0].set_ylabel('y')
    axes[0].set_title('Diagrama de fases')
    axes[0].legend()
    axes[0].grid()
    
    # dx/dt vs t
    axes[1].plot(t, dx_dt, label='dx/dt', color='r')
    axes[1].set_xlabel('t')
    axes[1].set_ylabel('dx/dt')
    axes[1].set_title('dx/dt vs t')
    axes[1].grid()
    
    # dy/dt vs t
    axes[2].plot(t, dy_dt, label='dy/dt', color='g')
    axes[2].set_xlabel('t')
    axes[2].set_ylabel('dy/dt')
    axes[2].set_title('dy/dt vs t')
    axes[2].grid()
    
    plt.tight_layout()
    plt.show()

def actualizar_interactivo(a, b, c, d):
    mostrar_sistema_matricial(a, b, c, d)

a_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=1, description='a')
b_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=2, description='b')
c_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=3, description='c')
d_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=4, description='d')

ui = widgets.HBox([a_slider, b_slider, c_slider, d_slider])
interactive_plot = widgets.interactive_output(actualizar_interactivo, {'a': a_slider, 'b': b_slider, 'c': c_slider, 'd': d_slider})

display(widgets.VBox([ui, interactive_plot]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def mostrar_sistema_matricial(a, b, c, d):
    A = np.array([[a, b], [c, d]])
    print("Sistema de ecuaciones en formato matricial:")
    print("d/dt [ x ] = [ a  b ] [ x ]")
    print("      [ y ]   [ c  d ] [ y ]")
    print("Matriz de coeficientes:")
    print(A)
    
    print("Sistema de ecuaciones diferenciales:")
    print(f"dx/dt = {a}x + {b}y")
    print(f"dy/dt = {c}x + {d}y")
    
    print("Ecuación característica para los valores propios:")
    print("|A - λI| = 0")
    print(f"| {a} - λ  {b} |")
    print(f"| {c}   {d} - λ | = 0")
    
    caracteristica = f"λ² - ({a} + {d})λ + ({a}*{d} - {b}*{c}) = 0"
    print("Desarrollo de la ecuación característica:")
    print(caracteristica)
    
    # Cálculo de los valores propios
    discriminante = (a + d)**2 - 4 * (a*d - b*c)
    if abs(discriminante) < 1e-20:
        discriminante = 0
    
    if discriminante > 0:
        lambda1 = ((a + d) + np.sqrt(discriminante)) / 2
        lambda2 = ((a + d) - np.sqrt(discriminante)) / 2
    elif discriminante == 0:
        lambda1 = lambda2 = (a + d) / 2
    else:
        lambda1 = complex((a + d) / 2, np.sqrt(-discriminante) / 2)
        lambda2 = complex((a + d) / 2, -np.sqrt(-discriminante) / 2)
    
    print("Valores propios de la matriz A:")
    print(f"λ₁ = {lambda1}")
    print(f"λ₂ = {lambda2}")
    
    # Graficar
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Diagrama de fases
    x_vals = np.linspace(-10, 10, 20)
    y_vals = np.linspace(-10, 10, 20)
    X, Y = np.meshgrid(x_vals, y_vals)
    U = a * X + b * Y
    V = c * X + d * Y
    
    axes[0].quiver(X, Y, U, V, color='blue')
    
    if np.iscomplex(lambda1) or np.iscomplex(lambda2):
        pass  # No mostrar vectores propios si los valores propios son imaginarios
    elif lambda1 == lambda2:
        axes[0].axvline(x=lambda1.real, color='r', linestyle='--', label=f"λ = {lambda1}")
    else:
        axes[0].axvline(x=lambda1.real, color='r', linestyle='--', label=f"λ₁ = {lambda1}")
        axes[0].axhline(y=lambda2.real, color='g', linestyle='--', label=f"λ₂ = {lambda2}")
    
    axes[0].set_xlabel('x')
    axes[0].set_ylabel('y')
    axes[0].set_title('Diagrama de fases')
    axes[0].legend()
    axes[0].grid()
    
    # dx/dt vs t
    t = np.linspace(0, 10, 100)
    dx_dt = np.exp(lambda1.real * t)
    dy_dt = np.exp(lambda2.real * t)
    
    axes[1].plot(t, dx_dt, label='dx/dt', color='r')
    axes[1].set_xlabel('t')
    axes[1].set_ylabel('dx/dt')
    axes[1].set_title('dx/dt vs t')
    axes[1].grid()
    
    # dy/dt vs t
    axes[2].plot(t, dy_dt, label='dy/dt', color='g')
    axes[2].set_xlabel('t')
    axes[2].set_ylabel('dy/dt')
    axes[2].set_title('dy/dt vs t')
    axes[2].grid()
    
    plt.tight_layout()
    plt.show()

def actualizar_interactivo(a, b, c, d):
    mostrar_sistema_matricial(a, b, c, d)

a_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=1, description='a')
b_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=2, description='b')
c_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=3, description='c')
d_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=4, description='d')

a_input = widgets.FloatText(value=1, description='a:')
b_input = widgets.FloatText(value=2, description='b:')
c_input = widgets.FloatText(value=3, description='c:')
d_input = widgets.FloatText(value=4, description='d:')

def actualizar_desde_input(a, b, c, d):
    a_slider.value = a
    b_slider.value = b
    c_slider.value = c
    d_slider.value = d

input_ui = widgets.HBox([a_input, b_input, c_input, d_input])
slider_ui = widgets.HBox([a_slider, b_slider, c_slider, d_slider])
interactive_plot = widgets.interactive_output(actualizar_interactivo, {'a': a_slider, 'b': b_slider, 'c': c_slider, 'd': d_slider})
input_interactive = widgets.interactive_output(actualizar_desde_input, {'a': a_input, 'b': b_input, 'c': c_input, 'd': d_input})

display(widgets.VBox([input_ui, slider_ui, interactive_plot]))
```
